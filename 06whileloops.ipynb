```python
n=1

while n<=5:
  print(n)
  n+=1
```

```python
n=1

while n!=0:
  n=int(input('Enter a number (0 to stop) : '))
  sq=n*n
  print(f'square of {n} is {sq}')
```

```python
nm=''

while nm!='stop':
  nm=input('Enter name : ')
  if nm!='stop':
     print(f'welcome {nm} to python')
```

```python
n=1

while True:
  if n==5:
    break
  print(n)
  n+=1
```
