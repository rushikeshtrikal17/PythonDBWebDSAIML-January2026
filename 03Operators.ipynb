Arithmetic Operators




```python
x=2
y=36

# arithmetic operators demo
print(x+y)
print(x-y)
print(x*y)
print(x/y)
# floor division
print(y//x)
#modulus
print(y%x)
print(27%4)
#power
print(7**2)
print(2**3)
```

Relational Operators

```python
a=5

print(a==3)
print(a==5)
password="rushikesh"
print(password=="trikal")

x=10
print(x>15)
print(x<15)
print(x<=10)
print(x>=15)
print(x!=7)
```

Assignment Operators

```python
x=9
x+=3    # x=x+3
x-=2
x*=5
x/=2
x//=5
x%=2
print(x)
```

Logical Operators

```python
print(10>5)
print(7<5)

print(10>5 and 7<5)
print(10>5 or 7<5)
print(not 9>13)
```

Membership operators

```python
print('p' in 'rushikesh')
print('p' not in 'rushikesh')

lst=['soham','praffull','sharayu','tejas','vijay']
print('megha' in lst)
```

Ternary Operator

```python
marks=78
result='pass' if marks>=35 else "fail"
print(result)
```
